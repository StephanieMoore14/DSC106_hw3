```python
def dct_unpack(orig):
    to_del = []
    to_add = {}
    for k, v in orig.items():
        if type(v) == dict:
            to_del.append(k)
            # unpack
            for k2, v2 in v.items(): 
                to_add[k2] = v2
    # delete old
    for key in to_del:
        del orig[key]
    orig.update(to_add)
    return orig
```

```python
gener = [i for i in total if i['id'] in gen]
prices = [i for i in total if i['id'] == 'Springfield.price']
temp = [i for i in total if i['id'] == 'Springfield.temperature']

# gen = {}
# for d in gener:
#     gen.update(d)
#gen = dct_unpack(gen)
gener
```
